```python
from ws_websocket_lib import KucoinCandlestickWebSocket
import pandas as pd
import json
from datetime import datetime,timezone
import importlib
import kucoin_fetch_spot
importlib.reload(kucoin_fetch_spot)
from kucoin_fetch_spot import KuCoinDataFetcher


print("imports done")

symbol = "BTC-USDT"
timeframe = "1min"
market_type = "spot"
start_time = "2025-01-07 00:00:00"

# Use timezone-aware objects to represent datetimes in UTC
end_time = datetime.now(timezone.utc).strftime("%Y-%m-%d %H:%M:%S")

fetcher = KuCoinDataFetcher(symbol, timeframe, market_type, start_time, end_time)

ws = KucoinCandlestickWebSocket(symbol, timeframe)

candle_list = fetcher._fetch_all_candles()

try:
    ws.start()
    
    while True:
        data = ws.get_data(timeout=1.0)
        if data:
            # Get candle data as list
            candle = data['candles']
            c= candle
            open = c[1]
            close = c[2]
            high = c[3]
            low = c[4]
            
            if candle[0] != candle_list[-1][0]:
                print('new candle printing update')
                # Append new candle to list
                candle_list.append(candle) 
                # makesure the list is not too long 
                if len(candle_list) > 500:
                    candle_list = candle_list[-500:]

                # Create DataFrame from single candle (wrap in list for 2D structure)
                df = pd.DataFrame(candle_list[-100:], columns=['timestamp', 'open', 'close', 'high', 'low', 'volume', 'turnover'])
                df['timestamp'] = pd.to_datetime(df['timestamp'].astype(int), unit='s', utc=True)
                df.drop(columns=['volume', 'turnover'], inplace=True)
                print(df)
            else:
                # Update last candle in list
                candle_list[-1] = candle
                df_temp = pd.DataFrame([candle_list[-1]], columns=['timestamp', 'open', 'close', 'high', 'low', 'volume', 'turnover'])
                df_temp['timestamp'] = pd.to_datetime(df_temp['timestamp'].astype(int), unit='s', utc=True)
                df_temp.drop(columns=['volume', 'turnover'], inplace=True)
                # print(df_temp)
            
except KeyboardInterrupt:
    ws.stop()
```
